```python
import requests
from bs4 import BeautifulSoup
```

```python
page=2
respond= requests.get(f'https://books.toscrape.com/catalogue/page-{page}.html')
respond
```

```python
soup = BeautifulSoup(respond.text,'lxml')
```

```python
def get_details(page_index, collection_books, index, books):
  #name of book
  title = collection_books[index].find('h3').find('a').text.strip('')
  #price
  price = collection_books[index].find('p',{'class':'price_color'}).text
  price = price[1:]
  #availability
  availability= collection_books[index].find('p',{'class':'instock availability'}).text.strip()
  # image url
  url= collection_books[index].find('a').find('img').get('src')
  books.append({'Page':page_index,'Title':title , 'Price':price, 'Availability':availability, 'Image_url':url })
```

```python
books = []
for i in range(1,3):
  respond= requests.get(f'https://books.toscrape.com/catalogue/page-{i}.html')
  collection = soup.find_all('li',{'class':'col-xs-6 col-sm-4 col-md-3 col-lg-3'})
  num_of_books = len(collection)
  print(i)
  for book_index in range(num_of_books):
    get_details(i,collection,book_index,books)
```
